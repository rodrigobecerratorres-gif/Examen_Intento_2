## **Verificar y corregir datos**

```python
print('Tipos ANTES de limpiar:')
print(df.dtypes)
print()

# Validacion de valores numericos
# --- critic_score: convertir a numerico ---
df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce')

# --- total_sales: ya deberia ser float, forzar conversion ---
df['total_sales'] = pd.to_numeric(df['total_sales'], errors='coerce')

# --- na_sales: ya deberia ser float, forzar conversion ---
df['na_sales'] = pd.to_numeric(df['na_sales'], errors='coerce')

# --- jp_sales: ya deberia ser float, forzar conversion ---
df['jp_sales'] = pd.to_numeric(df['jp_sales'], errors='coerce')

# --- pal_sales: ya deberia ser float, forzar conversion ---
df['pal_sales'] = pd.to_numeric(df['pal_sales'], errors='coerce')

# --- other_sales: ya deberia ser float, forzar conversion ---
df['other_sales'] = pd.to_numeric(df['other_sales'], errors='coerce')


# Convertir los campos tipo textos a fechas
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['last_update'] = pd.to_datetime(df['last_update'], errors='coerce')


print('Tipos DESPUES de limpiar:')
print(df.dtypes)
```

## **Tratamiento de Valores Nulos**

```python
# Verificar valores nulos
#Vaslores Nulos antes :
print('Valores Nulos antes :')
print('*'*40)
print(df.isnull().sum())
print('*'*40)
print('Total Filas antes del cambio:')
print(len(df))
print('*'*40)
# Columnas numéricas
columnas_numericas = ['total_sales','na_sales', 'jp_sales','pal_sales','other_sales']
columnas_categoricas = ['console','genre', 'publisher', 'developer']

# Criterio para el campo critic_score =0 para los nulos, que no tuvo calificaciones ese juego

df['critic_score'] = df['critic_score'].fillna(0)

# Criterio: Reemplazar nulos numéricos con mediana para las ventas
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].median())

# Criterio: Eliminar las filas que sus colunmas categoricas esten valores nulos para un registro
df = df.dropna(  subset=columnas_categoricas)

# Criterio: Las Columnas categoricas sin informacion se actualizara a ='Sin_info'
for col in columnas_categoricas:
    df[col] = df[col].fillna('Sin_info')

# Criterio: Actualizar los campos fecha con el valor ='2000-01-01'

df['release_date'] = df['release_date'].fillna(pd.Timestamp('2000-01-01'))
df['last_update'] = df['last_update'].fillna(pd.Timestamp('2000-01-01'))


# Verificar resultado
print(df.isnull().sum())


# Validar nuevamente
print('*'*40)
print('Valores Nulos Despues del cambio :')
print('*'*40)
print(df.isnull().sum())
print('*'*40)
print('Total Filas despues del cambio:')
print(len(df))


```

## **Visualizaciones exploratorias univariadas:**

**Crea dos tipos diferentes de visualizaciones univariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.hist(df['total_sales'], bins=30, edgecolor='black', alpha=0.7)

plt.title('Distribución de Ventas Totales')
plt.xlabel('Ventas Totales')
plt.ylabel('Frecuencia')

plt.show()
```

```python
#Graficar las Ventas por Genero del VideoJ uego
ventas_genero = (df.groupby('genre')['total_sales'].sum().sort_values(ascending=False))

plt.figure(figsize=(12,6))

ventas_genero.plot(kind='bar',  edgecolor='red', alpha=0.8)

plt.title('Ventas Totales por Genero')
plt.xlabel('Genero')
plt.ylabel('Ventas Totales')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Los juego de Accion y deportes son los que tienen mas ventas

```python
# Los juegos vendidos segun la critica obtenida segun su Puntajes de evaluacion

plt.figure(figsize=(10,6))

plt.scatter( df['critic_score'], df['total_sales'], alpha=0.5)

plt.title('Puntajes vs Ventas Totales')
plt.xlabel('Puntaje Evaluado')
plt.ylabel('Ventas Totales')

plt.grid(True)

plt.show()
```

## Visualizaciones exploratorias multivariadas:

**Crea dos tipos diferentes de visualizaciones multivariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.**

```python
df['año'] = df['release_date'].dt.year
ventas_Total = ( df.groupby('año')[['na_sales', 'jp_sales', 'pal_sales', 'other_sales'] ] .sum())
```

```python
#Ventas regional por año


ventas_anuales = ( df.groupby('año')[['na_sales', 'jp_sales', 'pal_sales', 'other_sales'] ] .sum())

print(ventas_anuales.head(5))

plt.figure(figsize=(16,7))

plt.plot( ventas_anuales.index, ventas_anuales['na_sales'], marker='o', linestyle='-', label='NA Sales')

plt.plot( ventas_anuales.index, ventas_anuales['jp_sales'], marker='s', linestyle='--', label='JP Sales')

plt.plot(ventas_anuales.index, ventas_anuales['pal_sales'], marker='^', linestyle='-.', label='PAL Sales')

plt.plot(ventas_anuales.index, ventas_anuales['other_sales'], marker='d', linestyle=':', label='Other Sales')


plt.title('Ventas Regionales por Año')
plt.xlabel('Año')
plt.ylabel('Ventas Totales')


plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)

plt.show()
```

Del grafico se puede interpretar que Norteamerica lleva la mayor venta.
- Hay tendencias bien marcadas en ciertos años que puede deberse a lanzamiento de nuevas plataformas.
- Por otro lado hay una baja despues del año 2020 por los juegos en linea.


```python
#Ventas de consolas por año

ventas_consola = (df.groupby(['año', 'console'])['total_sales'].sum().reset_index())

# Se filtro las 10 consutltas con mayor venta para graficarlo
top_consolas = (ventas_consola.groupby('console')['total_sales'].sum().sort_values(ascending=False).head(10).index)

ventas_top = ventas_consola[ventas_consola['console'].isin(top_consolas)]

print(ventas_top.head(5))


plt.figure(figsize=(15,8))

# Crear línea por consola
for consola in top_consolas:

    datos = ventas_top[ventas_top['console'] == consola]

    plt.plot(datos['año'], datos['total_sales'], marker='o', linewidth=2, label=consola )


plt.title('Ventas de Consolas por Año')
plt.xlabel('Año')
plt.ylabel('Ventas Totales')

plt.legend(title='Consolas', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.4)

plt.show()
```

Se puede apreciar que a partir del año 2000 hay un crecimiento en la venta de consolas.

Hay un reflejo de la evolucion de cada consola como salida al mercado, peak de ventas y su caida por obsolesencia.

Se puede apreciar las tendencias por plataforma/marca del mercado
